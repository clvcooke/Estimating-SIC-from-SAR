```python
network_configs = {
    'standard': [64, 64],
    'deeper': [64, 64, 64],
    'thin and deep': [32, 32, 32, 32, 32, 32, 32, 32, 32],
    'deep deep': [16, 16, 16, 16, 16, 16, 16]
}
for name, weights in network_configs.items():
    print(name)
    inp = keras.layers.Input((8,))
    lay = inp
    for weight in weights:
        lay = keras.layers.Dense(weight)(lay)
    lay = keras.layers.Dense(2)(lay)
    out = lay
    model = keras.Model(inp, lay)
    model.summary()
#     break
    
```
